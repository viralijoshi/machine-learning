```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv('/content/smaple.csv')
df






```

```python
plt.scatter(df['Current SPI'], df['iq level '] ,c=df['status '])
plt.show()


```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
print(x)
print(y)
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)#once you write fit in train you do not need to write it again in test
print(x_train)
print(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred =clf.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x_train,y_train.values,clf=clf)
plt.show()
```
